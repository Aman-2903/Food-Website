```python
def recommend(courses):
    index = course[course['Course Name'] == courses].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])  #to maintain index position
    L=[]
    for i in distances[1:6]:
        L.append(course.iloc[i[0]]['Course Name'])
        print(course.iloc[i[0]]['Course Name'])
    recommended_path.append(L)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

def evaluate_rmse_mae(actual_paths, recommended_paths):
    all_rmse = []
    all_mae = []

    for actual_path in actual_paths:
        path_rmse = []
        path_mae = []

        for recommended_path in recommended_paths:
            # Calculate RMSE and MAE for each pair of actual and recommended paths
            rmse = np.sqrt(mean_squared_error([0.5 if course in recommended_path else 1 for course in actual_path],
                                               [0.5 if course in actual_path else 1 for course in recommended_path]))
            mae = mean_absolute_error([0.5 if course in recommended_path else 1 for course in actual_path],
                                      [0.5 if course in actual_path else 1 for course in recommended_path])
            path_rmse.append(rmse)
            path_mae.append(mae)

        # Find the minimum RMSE and MAE for this actual path
        min_rmse = max(path_rmse)
        min_mae = max(path_mae)

        # Append the minimum values to the overall lists
        all_rmse.append(min_rmse)
        all_mae.append(min_mae)

    # Calculate the average RMSE and MAE across all actual paths
    avg_rmse = np.mean(all_rmse)
    avg_mae = np.mean(all_mae)

    return avg_rmse, avg_mae


rmse, mae = evaluate_rmse_mae(actual_recommendation, recommended_path)
print("RMSE:", rmse)
print("MAE:", mae)
```
